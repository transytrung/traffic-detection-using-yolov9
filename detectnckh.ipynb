```python
import cv2
import torch
from torchvision import transforms
from google.colab.patches import cv2_imshow
!pip install model
!pip install opencv-python-headless
!pip install ipython
!pip install matplotlib-venn
!apt-get -qq install -y libfluidsynth1
!apt-get -qq install -y graphviz && pip install pydot
import pydot
!pip install cartopy
import cartopy
# ƒê∆∞·ªùng d·∫´n ƒë·∫øn pre-trained model v√† script
model_path = '/content/drive/MyDrive/nckh/yolov9/runs/train/datrainxong23/weights/best.pt'
script_path = '/content/drive/MyDrive/nckh/yolov9/models/yolo.py'

# Import model class from script
from model import yolov9

# Load pre-trained model
model = yolov9()
model.load_state_dict(torch.load(model_path))
model.eval()

# Define transformations
transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])

# Open webcam
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Convert frame to tensor
    img_tensor = transform(frame).unsqueeze(0)

    # Predict
    with torch.no_grad():
        output = model(img_tensor)

    # Process output (you need to adjust this part according to your model)
    # Here's a simple example assuming your model outputs class probabilities
    _, predicted_class = torch.max(output, 1)
    class_index = predicted_class.item()

    # Display result
    cv2.putText(frame, f'Class: {class_index}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2_imshow(frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release webcam
cap.release()
```

```python
import cv2
import numpy as np
from IPython.display import display, HTML
from base64 import b64encode

# T·∫£i video t·ª´ Google Drive
from google.colab import drive
drive.mount('/content/drive')

# ƒê∆∞·ªùng d·∫´n ƒë·∫øn video ƒë·∫ßu v√†o v√† video ƒë·∫ßu ra
video_input_path = '/content/drive/MyDrive/nckh/yolov9/runs/detect/exp4/input.mp4'
video_output_path = '/content/drive/MyDrive/nckh/yolov9/runs/detect/exp4/output_detected.mp4'

# Load m√¥ h√¨nh nh·∫≠n di·ªán
net = cv2.dnn.readNet('/content/drive/MyDrive/nckh/yolov9/runs/train/datrainxong23/weights/best.pt', '/content/drive/MyDrive/nckh/yolov9/yolov9.cfg')

# ƒê·ªçc video ƒë·∫ßu v√†o
cap = cv2.VideoCapture(video_input_path)

# X√°c ƒë·ªãnh c√°c l·ªõp v√† m√†u s·∫Øc
classes = ['Rickshaw', 'Tractor', 'bus', 'car', 'motorcycle', 'plate no-', 'truck', 'van']
colors = [(255, 0, 0), (0, 255, 0), (0, 0, 255), (255, 255, 0), (255, 0, 255), (0, 255, 255), (128, 0, 0), (0, 128, 0)]

# L·∫•y th√¥ng s·ªë video ƒë·∫ßu v√†o
fps = cap.get(cv2.CAP_PROP_FPS)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# T·∫°o video ƒë·∫ßu ra
out = cv2.VideoWriter(video_output_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (frame_width, frame_height))

# V√≤ng l·∫∑p qua c√°c frame
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Chuy·ªÉn ƒë·ªïi frame th√†nh blob ƒë·ªÉ ƒë∆∞a v√†o m·∫°ng
    blob = cv2.dnn.blobFromImage(frame, 1 / 255.0, (416, 416), swapRB=True, crop=False)

    # ƒê∆∞a blob v√†o m·∫°ng v√† nh·∫≠n c√°c output
    net.setInput(blob)
    layer_outputs = net.forward()

    # Duy·ªát qua c√°c output t·ª´ c√°c l·ªõp yolo
    for output in layer_outputs:
        for detection in output:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]

            # N·∫øu ƒë·ªô tin c·∫≠y l·ªõn h∆°n ng∆∞·ª°ng
            if confidence > 0.5:
                # T√≠nh to√°n bounding box
                box = detection[0:4] * np.array([frame.shape[1], frame.shape[0], frame.shape[1], frame.shape[0]])
                (center_x, center_y, width, height) = box.astype("int")
                x = int(center_x - (width / 2))
                y = int(center_y - (height / 2))

                # V·∫Ω bounding box v√† hi·ªÉn th·ªã nh√£n
                cv2.rectangle(frame, (x, y), (x + width, y + height), colors[class_id], 2)
                cv2.putText(frame, classes[class_id], (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, colors[class_id], 2)

    # Ghi frame ƒë√£ x·ª≠ l√Ω v√†o video ƒë·∫ßu ra
    out.write(frame)

# Gi·∫£i ph√≥ng c√°c t√†i nguy√™n
cap.release()
out.release()
cv2.destroyAllWindows()

# Hi·ªÉn th·ªã video ƒë·∫ßu ra trong notebook
output_video = open(video_output_path, "rb").read()
encoded_video = b64encode(output_video).decode()
video_html = f"""<video width="640" height="480" controls><source src="data:video/mp4;base64,{encoded_video}" type="video/mp4"></video>"""
display(HTML(video_html))
```
